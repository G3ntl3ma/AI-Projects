```python
"""
Klassifizierung der Appendix-Datensätze mit mehrschichtigem Netz 15x8x1 ohne Aktivierung
"""
import numpy as np
from infile import readfile

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def scale(liste):
    #Eingabewerte anpassen
    for x in liste:
        x[0] = x[0] / 80; # Alter
        x[1] = x[1] - 0.9; # m/w                              
        x[11] = ( x[11]-36.0 )/4; #Temperatur
        x[12] = ( x[12]-36.0 )/4;
        x[13] = x[13]/24000.0; #Leukozyten
        
        for i in range(15):
            if x[i] ==0:
                x[i] = x[i] + 0.1


def output(vektor):
  #Berechnen des Ausgabewertes y

  y = 0 #Zurücksetzen
  inputs = np.zeros([15])
 
  # letzten Wert "abschneiden"  
  for i in range( 15 ):
      inputs[i] = vektor[i]
  
  # Eingabevektor mal Gewichtsmatrix = mittlere Schicht hL
  global hL
  hL = np.dot(inputs,wil)
    
  oL = [] #Output

  oL = np.dot(hL,wol)
  
  print(oL[0])
  y = oL[0]

  return y

def train():
    # Ein Durchlauf mit Fehlerrückführung
    #Zielwert
    t = x[15]
    
    #Anwenden der Delta_regel für die Outputschicht
    for i in range(4):
        hL[i] = sigmoid(hL[i])
        delta = eta * hL[i] * ( t - y ) * y * (1 - y)
        wol[i] = wol[i] + delta

    
    
    #Berechnung des Fehlers
    error=t-y
    
    E=0
    for i in range(len(wol)):
        E=E+wol[i]
        
    #Aufteilung des Fehlers
    e=np.zeros([len(hL)])
    for i in range(len(hL)):
        e[i]=wol[i]/E*error
    
    #Gewichtsanpassung für die erste Schicht
    for j in range(8):
        for i in range(15):
            
            delta=eta*x[i]*e[j]*hL[j] *(1-hL[j])
            
            wil[i][j]=delta+wil[i][j] 
            
"""
Hauptteil
"""

# Gewichtematrix InputLayer 15x4, Gewichte OutputLayer
# Füllen mit Zufallszahlen
wil=np.random.rand(15,8)-0.5
wol=np.random.rand(8,1) - 0.5

#print(wil)
#print(wol)

#Testdatei einlesen
testListe = readfile("App157.dat")
print( "Testdatei mit ", len(testListe)," Zeilen eingelesen" )

scale(testListe)

#Trainingsdatei einlesen
trainListe = readfile("App316.dat")
print( "Trainingsdatei mit ",len(trainListe)," Zeilen eingelesen" )

scale(trainListe)

#Lernrate
eta = 0.3; iterations =1 ;

print("\nTraining!\n")

for i in range(iterations):
    for x in trainListe:
        y = output(x) # y neu berechnen 
        
        # Trainieren des Netzes mit der Delta-Regel
        train()

#Netzausgabe berechnen

result=[]

for x in testListe:
    
    y=output(x)
    
    print(y)
     
    #Ergebnisliste
    result.append(round(y))

#Auswertung
r = 0
for i in range( len(testListe) ):
    if testListe[i][15]==result[i]:
       r = r+1 
    
print("\nRichtig: ",r, " das sind ", round( r/len(testListe) * 100), "%" )

```
